```python
# loads data
air_temp = np.loadtxt('/Users/hgducharme/Programming/iPython/notebook1/holland_temperature.dat')
sea_temp = np.loadtxt('/Users/hgducharme/Programming/iPython/notebook1/holland_seawater.dat')

# creates 2 subplots, sets x axis, and defines x axis labels
f, (ax1, ax2) = plt.subplots(2, sharex=True)
f.canvas.draw()
x = np.linspace(1,12,12)
months = ["Jan", "Feb", "Mar", "Apr", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# creates 1st subplot 
ax1.plot(x, air_temp, 'r-', x, sea_temp, 'b-')
ax1.set_title("Holland's Sea and Air Temperatures")
ax1.set_ylabel("Mean Monthly Temperatures (Celcius)")
ax1.legend(["Air Temperature", "Sea Temperature"], bbox_to_anchor=(1.5,1))

# creates 2cd subplot
temp_difference = []
(air_temp - sea_temp).all(temp_difference.append(air_temp - sea_temp))       
for i in temp_difference:
    ax2.plot(x, i, 'ko')
ax2.set_ylabel("Air - Sea Temperature (Celcius)")
ax2.set_xlabel("Month of the Year")

# defines common x-labels
ax1,ax2.set_xticks(range(1,13));
ax1,ax2.set_xticklabels(months);
```

```python
air_temp = np.loadtxt('/Users/hgducharme/Programming/iPython/notebook1/holland_temperature.dat')
sea_temp = np.loadtxt('/Users/hgducharme/Programming/iPython/notebook1/holland_seawater.dat')

months = ["Jan", "Feb", "Mar", "Apr", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
x = np.linspace(1,12,12, months)

plt.fill_between(x, air_temp, color='red', alpha=.3)
plt.fill_between(x, sea_temp, color= 'blue', alpha=.3)

plt.title("Holland's Air and Seawater Temperatures", fontsize='small')
plt.xlabel('Month', fontsize="small")
plt.ylabel('Monthly Temperatures (Celcius)', fontsize="small")
plt.xticks(range(1,13), months, fontsize='small');
plt.tight_layout()
```
